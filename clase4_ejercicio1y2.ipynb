```python
def log_plot(FactorEscalado):
    x = np.arange(0, 256, 0.01)
    y = FactorEscalado * np.log(1 + x) *10
    plt.plot(x, y, 'r', linewidth=1)
    #plt.rcParams['font.sans-serif']=['SimHei'] 
    plt.title(u'grafica logaritmo')
    plt.xlim(0, 255), plt.ylim(0, 255)
    plt.show()
```


> $$
output = c log[1+e^escalado -1)* r]
$$
> $$
c = MaxR / [log [1+e^sigma - 1) * maxr]
$$


```python
def log(FactorEscalado, img):
    # https://stackoverflow.com/questions/21117415/finding-the-value-of-the-min-and-max-pixel
    smallest = img.min(axis=(0, 1))
    largest = img.max(axis=(0, 1))
    
    print(smallest,largest)
    
    maxS = 255
    maxR = largest

    c = maxS/(np.log(1+(math.exp(FactorEscalado)-1)* maxR))

    output  = c * np.log( 1 + ((math.exp(FactorEscalado) - 1) * img))
        
    output = np.uint8(output)
    #log_plot(img, output
    return output
```

```python
FactorEscalado = int(input("Introduce factor de escalado: "))
```

```python
print("El factor de escalado es:", FactorEscalado)
```

```python
img = cv2.imread('img/food.jpg')

smallest = img.min(axis=(0, 1))
largest = img.max(axis=(0, 1))
    
print(smallest,largest)
output = log(FactorEscalado, img)
```

Clase 4 Ejercicio 2. Implementar un programa Python que aplique una transformaciÃ³n exponencial a una imagen sobrepuesta de entrada en color con factor de escalado parametrizable

```python
FactorEscalado = int(input("Introduce factor de escalado: "))
```

```python
print("El factor de escalado es:", FactorEscalado)
FactorEscalado = 100
```

```python
def expo(FactorEscalado, img):
    # https://stackoverflow.com/questions/21117415/finding-the-value-of-the-min-and-max-pixel
    smallest = img.min(axis=(0, 1))
    largest = img.max(axis=(0, 1))
    
    print(smallest,largest)
    
    maxS = 255
    maxR = largest

    c = maxS / ((1 + FactorEscalado)^img) - 1

    output  = c * [((1 + FactorEscalado)^maxR)  -1]
        
    output = np.uint8(output)
    #log_plot(img, output
    return output
```

```python
img = cv2.imread('img/food.jpg')

smallest = img.min(axis=(0, 1))
largest = img.max(axis=(0, 1))
    
print(smallest,largest)
output = expo(FactorEscalado, img)
```
